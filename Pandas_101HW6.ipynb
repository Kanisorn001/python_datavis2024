````
# 1. นิทานที่ยาวที่สุดและสั้นที่สุด (ไม่นับสัญลักษณ์พิเศษ)
folk_df['description_length'] = folk_df['description'].astype(str).apply(lambda x: len(''.join(c for c in x if c.isalnum())))  # Convert the 'description' column to string type before applying the lambda function
longest_index = folk_df['description_length'].idxmax()
shortest_index = folk_df['description_length'].idxmin()
longest_title = folk_df.loc[longest_index, 'description_length']
shortest_title = folk_df.loc[shortest_index, 'description_length']

print(f'นิทานที่ยาวที่สุด = {folk_df.loc[longest_index, "title"]} (จำนวน {longest_title} ตัวอักษร)')
print(f'นิทานที่สั้นที่สุด = {folk_df.loc[shortest_index, "title"]} (จำนวน {shortest_title} ตัวอักษร)')

# 2. ชนิดของนิทาน (type) ที่มากที่สุดและน้อยที่สุด
type_counts = folk_df['type'].value_counts()
most_common_type = type_counts.idxmax()
least_common_type = type_counts.idxmin()

print("ชนิดของนิทานที่มากที่สุด:", most_common_type)
print("ชนิดของนิทานที่น้อยที่สุด:", least_common_type)

# 3. มี spatial ที่ไม่ใช่ของไทยกี่เรื่อง
non_thai_spatial_count = folk_df[folk_df['spatial'] != 'ไทย'].shape[0]

print("จำนวนเรื่องที่มี spatial ไม่ใช่ของไทย:", non_thai_spatial_count)
````

##HW6
บรรยายเกี่ยวกับผู้กำกับชั้นนำของอินเดีย 3 คน ว่าทำไมเราถึงคิดว่าเป็นผู้กำกับชั้นนำ และชั้นนำยังไง

````
import pandas as pd
````

````
india = pd.read_csv('IMDb Movies India.csv',encoding = 'latin-1')
india
````

````
# นับจำนวนหนังที่แต่ละผู้กำกับทำ
director_counts = india['Director'].value_counts().reset_index()
director_counts.columns = ['Director', 'MovieCount']

# คำนวณค่าเฉลี่ยของจำนวนหนังที่ผู้กำกับหนึ่งคนทำ
average_movies_per_director = director_counts['MovieCount'].mean()

print(f"ค่าเฉลี่ยว่าผู้กำกับหนึ่งคนทำหนังกี่เรื่อง: {average_movies_per_director}")
````

````


# ตรวจสอบว่ามีคอลัมน์ที่จำเป็นหรือไม่
required_columns = ['Director', 'Rating', 'Votes']
if not all(column in india.columns for column in required_columns):
    raise ValueError("ไฟล์ข้อมูลไม่มีคอลัมน์ที่จำเป็น (Director, Rating, Votes)")

# กรองข้อมูลที่มีการให้คะแนนและโหวต
data = india.dropna(subset=['Rating', 'Votes'])

# แปลงคอลัมน์ 'Rating' และ 'Votes' เป็นตัวเลข
# errors='coerce' จะแทนค่าที่ไม่สามารถแปลงเป็นตัวเลขได้ด้วย NaN
data['Rating'] = pd.to_numeric(data['Rating'], errors='coerce')
data['Votes'] = pd.to_numeric(data['Votes'], errors='coerce')

# กรองเฉพาะผู้กำกับที่ทำหนังมากกว่า 3 เรื่อง
director_counts = data['Director'].value_counts().reset_index() # Calculate director counts
director_counts.columns = ['Director', 'MovieCount']
directors_more_than_3_movies = director_counts[director_counts['MovieCount'] > 3]

# ลบแถวที่มีค่า NaN ในคอลัมน์ 'Rating' หรือ 'Votes'
data = data.dropna(subset=['Rating', 'Votes'])

# คำนวณคะแนนถ่วงน้ำหนัก (weighted average rating) for the main DataFrame
data['WeightedRating'] = data['Rating'] * data['Votes']

# คำนวณค่าเฉลี่ยการให้คะแนนและจำนวนโหวตสำหรับผู้กำกับที่ทำหนังมากกว่า 10 เรื่อง
director_stats = data[data['Director'].isin(directors_more_than_3_movies['Director'])]
director_stats = director_stats.groupby('Director').agg({'Rating': 'mean', 'Votes': 'sum', 'WeightedRating': 'sum'}).reset_index() # Include WeightedRating in aggregation

# คำนวณค่าเฉลี่ยถ่วงน้ำหนักสำหรับแต่ละผู้กำกับ
director_stats['WeightedAverageRating'] = director_stats['WeightedRating'] / director_stats['Votes']

# หาผู้กำกับที่มีค่าเฉลี่ยถ่วงน้ำหนักสูงสุด 3 อันดับ
top_directors = director_stats.sort_values(by='WeightedAverageRating', ascending=False).head(3)

print(top_directors[['Director', 'WeightedAverageRating']])
````
