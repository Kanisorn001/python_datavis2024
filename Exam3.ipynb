```python
import pandas as pd
import numpy as np
```

```python
Data = pd.read_csv('Doctor.csv')
Data
```

```python
Data1 = Data[Data['comments'].str.contains('มะเร็ง')]
Data1
```

```python
DataR = Data1.sort_values(by=['ยอดไลค์', 'จำนวนการตอบกลับ'], ascending=False)
DataR
```

```python
# prompt: Filter  ว่า comments ไหน ยาวที่สุด

DataR[DataR['comments'].str.len() == DataR['comments'].str.len().max()]
```

```python
import re

# รายชื่อชนิดมะเร็งที่ต้องการค้นหา
cancer_types = [
    'มะเร็งเต้านม', 'มะเร็งปอด', 'มะเร็งลำไส้', 'มะเร็งต่อมลูกหมาก','มะเร็งขั้วปอด',
    'มะเร็งกระเพาะอาหาร', 'มะเร็งตับ', 'มะเร็งตับอ่อน', 'มะเร็งรังไข่', 'มะเร็งผิวหนัง',
    'มะเร็งปากมดลูก', 'มะเร็งเม็ดเลือดขาว', 'มะเร็งต่อมน้ำเหลือง', 'มะเร็งกระดูก',
    'มะเร็งสมอง', 'มะเร็งไต', 'มะเร็งต่อมไทรอยด์', 'มะเร็งหลอดอาหาร', 'มะเร็งกระเพาะปัสสาวะ',
    'มะเร็งปาก', 'มะเร็งหลอดลม', 'มะเร็งตับอ่อน', 'มะเร็งไทรอยด์', 'มะเร็งเซลล์', 'มะเร็งต่อมหมวกไต',
    'มะเร็งต่อมปิเนียล', 'มะเร็งผิวหนัง', 'มะเร็งเซลล์ประสาท','มะเร็งเม็ดสีผิว','มะเร็งลำคอ','มะเร็งโพรงจมูก'
]

# ฟังก์ชันในการตรวจสอบว่ามีชนิดมะเร็งใดบ้างในคอมเมนต์
def find_cancer_mentions(comment):
    found_cancers = [cancer for cancer in cancer_types if cancer in comment]
    return found_cancers if found_cancers else None

# ค้นหามะเร็งในคอมเมนต์ทั้งหมด
comments = DataR['comments']
cancer_mentions = comments.apply(find_cancer_mentions).dropna()

# นับจำนวนคอมเมนต์ที่พบชนิดมะเร็ง
print(f"จำนวนคอมเมนต์ที่พบชนิดมะเร็ง:", len(cancer_mentions))
```

```python
# prompt: เพิ่ม column ชนิดมะเร็ง

# สร้าง column ใหม่ชื่อ 'ชนิดมะเร็ง' และเติมค่าเริ่มต้นเป็น None
DataR['ชนิดมะเร็ง'] = None

# วนลูปเพื่อเติมข้อมูลใน column 'ชนิดมะเร็ง'
for idx, cancers in cancer_mentions.items():
  DataR.at[idx, 'ชนิดมะเร็ง'] = ', '.join(cancers)
```

```python
# prompt: จัดอันดับว่าคนเป็น มะเร็งชนิดไหนมากที่สุด

from collections import Counter

# นับจำนวนชนิดมะเร็งทั้งหมด
all_cancers = [cancer for sublist in cancer_mentions.tolist() for cancer in sublist]
cancer_counts = Counter(all_cancers)

# แสดงผลลัพธ์การนับชนิดมะเร็ง
print("จำนวนผู้ป่วยมะเร็งแต่ละชนิด:")
for cancer, count in cancer_counts.most_common():
    print(f"{cancer}: {count} คน")
```
