```python
import pandas as pd

# Load the Excel file
file_path = '/content/test-TidyData_2024.xlsx'
excel_data = pd.ExcelFile(file_path)

# Load the data from the 'water_quality' sheet
data = pd.read_excel(file_path, sheet_name='water_quality')

##1. ความแตกต่างของ ค่าเฉลี่ย ของข้อมูล Chlo_a ทุกๆ station ระหว่างปี 2018 และ ปี 2019

# Extract year information from the 'month' column for easier comparison
data['year'] = data['month'].apply(lambda x: x.split('_')[1])

# Calculate the mean of 'Chlo_a' for each station and year (2018 and 2019)
mean_chlo_a = data.groupby(['station', 'year'])['Chlo_a'].mean().unstack()

# Calculate the difference between 2018 and 2019 for each station
mean_chlo_a['difference'] = mean_chlo_a['2019'] - mean_chlo_a['2018']

##2. สร้างตารางค่าเฉลี่ย Chlo_b ทุกๆ station ของแต่ละเดือน

# Create a pivot table to calculate the mean of 'Chlo_b' for each station and each month
data['month_only'] = data['month'].apply(lambda x: x.split('_')[0])
mean_chlo_b = data.pivot_table(values='Chlo_b', index='station', columns='month_only', aggfunc='mean')
```
